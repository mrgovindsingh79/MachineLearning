```python
import csv
import random
import math
import operator
from math import sqrt
import numpy as p
 
def loadDataset(filename, filename1, trainingSet=[] , testSet=[]):   #Loading training dataset
    with open(filename, 'rt') as csvfile:
        lines = csv.reader(csvfile)
        dataset = list(lines)
        for x in range(len(dataset)):
            for y in range(24):
                dataset[x][y] = int(dataset[x][y])   #converting the value into integer form
            
            trainingSet.append(dataset[x])
            
    with open(filename1) as csvfile:    #loading tes
            lines = csv.reader(csvfile)
            dataset = list(lines)
            for x in range(len(dataset)):
                for y in range(24):
                    dataset[x][y] = int(dataset[x][y])
                testSet.append(dataset[x])

 

 

 


```

```python
def euclideanDistance(instance1, instance2, length):
    distance = 0
    for x in range(length-1):
        distance += pow((int(instance1[x]) - int(instance2[x])), 2)
    return math.sqrt(distance)
```

```python
def getNeighbors(trainingSet, testInstance, k):
    distances = []
    length = len(testInstance)-1
    for x in range(len(trainingSet)):
        dist = euclideanDistance(testInstance, trainingSet[x], length)
        distances.append((trainingSet[x], dist))     #Append every distance calculated by the EuclideanDisatnce function
    distances.sort(key=operator.itemgetter(1))  #sort the distances
    neighbors = []
    for x in range(k):
        neighbors.append(distances[x][0])
    return neighbors          #returns the K most nearest distances
```

```python
def getResponse(neighbors):
    classVotes = {}
    for x in range(len(neighbors)):
        response = neighbors[x][-1]
        if response in classVotes:
            classVotes[response] += 1
        else:
            classVotes[response] = 1
    sortedVotes = sorted(classVotes.items(), key=operator.itemgetter(1), reverse=True)
    return sortedVotes[0][0]
```

```python
def getAccuracy(testSet, predictions):
    correct = 0
    for x in range(len(testSet)):
        if testSet[x][-1] == predictions[x]:
            correct += 1
    return (correct/float(len(testSet))) * 100.0
```
